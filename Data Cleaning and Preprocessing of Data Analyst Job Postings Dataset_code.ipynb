```python
import pandas as pd

df = pd.read_csv(r"DataAnalyst-Raw-Student-V2 (1).csv")
print(df.shape)        # Check number of rows & columns
df.head()              # Preview first few rows
df.info()              # See data types and null counts
df.describe()          # Summary statistics (only for numeric columns)
```

```python
#Handling Missing Value
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)

```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.dtypes
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['founded'] = pd.to_datetime(df['founded'], errors='coerce')
```

```python
df['company_name'] = df['company_name'].str.strip()
df['location'] = df['location'].str.title()
```

```python
import numpy as np
Q1 = df['rating'].quantile(0.25)
Q3 = df['rating'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['rating'] < (Q1 - 1.5 * IQR)) | (df['rating'] > (Q3 + 1.5 * IQR)))]
```

```python
df['type_of_ownership'] = df['type_of_ownership'].str.lower().str.replace(' ', '_')
```

```python
print(df.info())
print(df.isnull().sum())
df.head()
```
